# 1. Leitura de arquivos

```python
df1 = pd.read_excel('/content/commits_projetos_1.xlsx')
df2 = pd.read_excel('/content/commits_projetos_2.xlsx')
df3 = pd.read_excel('/content/commits_projetos_3.xlsx')
df4 = pd.read_csv('/content/commits_projetos_4.csv')
df5 = pd.read_csv('/content/commits_projetos_5.csv')
```

# 2. Merge de dados

Juntando todos os dados que temos de commits coletados

```python
all_data = pd.concat([df1, df2, df3, df4, df5])
all_data.shape
```

```python
all_data.shape
```

```python
all_data.head(1)
```

Removendo commits duplicados

```python
# considerando apenas 'sha' exclui-se repos 'forkados'
merged_data_df = all_data.drop_duplicates(subset=['sha', 'repo'], keep="last")
merged_data_df.shape
```

```python
merged_data_df.isna().sum()
```

# 3. Salvar merge de commits

```python
brasilia_tz = pytz.timezone('America/Sao_Paulo')
now = datetime.datetime.now(brasilia_tz).strftime("%d-%m-%Y_%H-%M-%S")
print(now)

merged_data_df.to_csv(f"merged_commits_{now}.csv", index=False)
```

```python
all_data.to_csv(f"all_commits_{now}.csv", index=False)
```
